```python
# Module importieren
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Random generiertes DataFrame erstellen
df = pd.DataFrame(np.random.randint(0,100,size=(1000, 2)), columns=list('XY'))

# Kopfzeile ausgeben
df.head(10)
```

```python
# DataFarme random selektieren fuer Webers-law Visualisierung


df_10 = df.sample(n=10, random_state= 1)
df_20 = df.sample(n=20, random_state= 1)
df_200 = df.sample(n=200, random_state= 1)
df_210 = df.sample(n=210, random_state= 1)
```

```python
# Scatterplot Visualisierungen

def scatterplot(df, title, colors):
    plt.figure(figsize=(15,5))
    sns.scatterplot(data = df, x = "X", y = "Y", s = 50, color = colors)
    plt.title(title, fontsize = 15)
    plt.show()
    
```

```python
# Visualisieren mit 10 Punkten
scatterplot(df_10, "10 Punkte", "b")
```

```python
# Visualisieren mit 20 Punkten
scatterplot(df_20, "20 Punkte", "b")
```

```python
# Visualisieren mit 200 Punkten
scatterplot(df_200, "200 Punkte", "orange")
```

```python
# Visualisieren mit 210 Punkten
scatterplot(df_210, "210 Punkte", "orange")
```
